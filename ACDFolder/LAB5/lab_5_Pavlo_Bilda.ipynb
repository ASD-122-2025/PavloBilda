```python
# Імпорт необхідних бібліотек
import time
import random
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Markdown

# 1. Реалізація алгоритмів пошуку

def linear_search(arr, x):
    """
    Лінійний пошук
    Складність: O(n)
    """
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1

def binary_search(arr, x):
    """
    Бінарний пошук (працює тільки на відсортованих масивах)
    Складність: O(log n)
    """
    low = 0
    high = len(arr) - 1
    
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1

# 2. Тестування алгоритмів

# Генеруємо тестові дані
sizes = [100, 1000, 10000, 100000]
linear_times = []
binary_times = []

for size in sizes:
    arr = sorted([random.randint(0, 100000) for _ in range(size)])
    x = random.choice(arr)
    
    # Вимірюємо час для лінійного пошуку
    start = time.time()
    linear_search(arr, x)
    linear_times.append(time.time() - start)
    
    # Вимірюємо час для бінарного пошуку
    start = time.time()
    binary_search(arr, x)
    binary_times.append(time.time() - start)

# 3. Візуалізація результатів
plt.figure(figsize=(10, 6))
plt.plot(sizes, linear_times, 'o-', label='Лінійний пошук (O(n))')
plt.plot(sizes, binary_times, 's-', label='Бінарний пошук (O(log n))')
plt.xlabel('Розмір масиву')
plt.ylabel('Час виконання (секунди)')
plt.title('Порівняння ефективності алгоритмів пошуку')
plt.legend()
plt.grid(True)
plt.show()

# 4. Контрольні питання
display(Markdown("""
### Контрольні питання:

1. **Яка асимптотична складність лінійного та бінарного пошуку?**  
   - Лінійний пошук: O(n)  
   - Бінарний пошук: O(log n)

2. **Чому бінарний пошук вимагає відсортованого масиву?**  
   Бінарний пошук використовує стратегію "розділяй і володарюй", порівнюючи шуканий елемент з елементом у середині масиву. 
   Це працює лише коли масив відсортований, оскільки ми можемо точно визначити, в якій половині масиву продовжувати пошук.

3. **Які ще алгоритми пошуку існують?**  
   - Інтерполяційний пошук (покращена версія бінарного)  
   - Пошук стрибками (Jump Search)  
   - Експоненційний пошук  
   - Пошук Фібоначчі
"""))
```
