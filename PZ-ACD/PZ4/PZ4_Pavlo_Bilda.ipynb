```python
import time
import matplotlib.pyplot as plt

def linear_search(lst, target):
    for i, val in enumerate(lst):
        if val == target:
            return i
    return -1

def binary_search(lst, target):
    left, right = 0, len(lst) - 1
    while left <= right:
        mid = (left + right) // 2
        if lst[mid] == target:
            return mid
        elif lst[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

sizes = [100, 1000, 5000, 10000]
lin_times = []
bin_times = []

for size in sizes:
    data = list(range(size))
    target = size - 1

    start = time.time()
    linear_search(data, target)
    lin_times.append(time.time() - start)

    start = time.time()
    binary_search(data, target)
    bin_times.append(time.time() - start)

plt.plot(sizes, lin_times, label="Лінійний пошук")
plt.plot(sizes, bin_times, label="Бінарний пошук")
plt.xlabel("Розмір списку")
plt.ylabel("Час виконання (сек)")
plt.title("Порівняння пошукових алгоритмів")
plt.legend()
plt.grid(True)
plt.show()
```
